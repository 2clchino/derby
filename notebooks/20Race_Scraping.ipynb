```python
PRED_RACE = 'id=0059'
```

```python
TRAIN_RACE_ID = func.get_train_race_id(func.get_text_from_page('https://race.netkeiba.com/special/index.html?' + PRED_RACE))
PRED_RACE_ID = func.get_pred_race_id(func.get_text_from_page('https://race.netkeiba.com/special/index.html?' + PRED_RACE))
```

```python
RACE = "/race/"
CSV_DIR = "csv"
URL_BASE = "https://db.netkeiba.com"
URL_RACE = "https://race.netkeiba.com/race/shutuba.html?race_id="
RACE_TABLE_NAME = "race_table_01 nk_tb_common"
```

### スクレイピング

```python
if __name__ == '__main__':
    url = URL_RACE + PRED_RACE_ID
    text = func.get_text_from_page(url)
    RACE_NAME = func.get_name_from_text(text).replace(' - netkeiba.com', '').split('(')[0]
    print(RACE_NAME)
    print(PRED_RACE_ID)
    if not os.path.exists(CSV_DIR + "/horse/" + PRED_RACE_ID[0:4] + RACE_NAME):
        os.makedirs(CSV_DIR + "/horse/" + PRED_RACE_ID[0:4] + RACE_NAME)
    (info, horse_id, horse_names) = func.get_race_info_from_text(text, "Shutuba_Table RaceTable01 ShutubaTable", PRED_RACE_ID[0:4], RACE_NAME)
    file_path = CSV_DIR+ RACE + PRED_RACE_ID[0:4] + RACE_NAME + ".csv"
    with open(file_path, "w", newline="", encoding='shift_jis') as f:
        writer = csv.writer(f)
        writer.writerows(info)
```

### 訓練データ取得

```python
def correct_train_data(race_id):
    url = URL_BASE + RACE + race_id
    text = func.get_text_from_page(url)
    print(race_id)
    
    if not os.path.exists(CSV_DIR + "/horse/" + race_id[0:4] + RACE_NAME):
        os.makedirs(CSV_DIR + "/horse/" + race_id[0:4] + RACE_NAME)
    if not os.path.exists(CSV_DIR + RACE + RACE_NAME):
        os.makedirs(CSV_DIR + RACE + RACE_NAME)
    (info, horse_id, horse_names) = func.get_old_race_info_from_text(False, text, RACE_TABLE_NAME, race_id, RACE_NAME)
    file_path = CSV_DIR+ RACE + RACE_NAME + "/" + race_id[0:4] + ".csv"
    with open(file_path, "w", newline="", encoding='shift_jis') as f:
        writer = csv.writer(f)
        writer.writerows(info)
```

```python
for i in range(len(TRAIN_RACE_ID)):
    correct_train_data(TRAIN_RACE_ID[i])
    time.sleep(3) # サーバーへの負荷を考慮してディレイを入れる。
```
